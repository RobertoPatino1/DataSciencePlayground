```python
import matplotlib.pyplot as plt
import numpy as np
#matplotlib suele trabajar en conjunto con numpy
#El modulo para graficar es pyplot
```

```python
#PodemosCrear graficas con esta libreria!
#Dibujando una recta de un punto a otro
coordenadasX = np.array([0,6])
coordenadasY = np.array([0,250])
```

```python
#la funcion plot dibuja una linea de punto a punto
plt.plot(coordenadasX,coordenadasY)
plt.show()
```

```python
#Otro grafico
#El primer argumento pasado a plot corresponde a los puntos en x, el segundo a los puntos en y
#OJO: El numero de elementos en las listas debe ser el mismo para ambos parametros
#You can plot as many points as you like, just make sure you have the same number of points in both axis.
plt.plot([1,2,3],[5,7,3])
plt.show()
```

Tambien podemos dibujar los puntos sin unirlos con una recta <br />
Para esto solo agregamos un argumento mas a la funcion plot <br /> 
plt.plot([puntos x],[puntos y],"o")

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16],"o")
```

En caso de pasarle solo un arreglo a la funcion plot, esta asumira que se trata de los puntos en "Y"<br />

```python
plt.plot([3, 8, 1, 10, 5, 7],"r") #Pasando el argumento r podemos definir el color de la recta
plt.show()
```
