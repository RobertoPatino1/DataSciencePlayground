```python
import matplotlib.pyplot as plt
import numpy as np
```

Al momento de usar plot podemos pasarle a la funcion un parametro llamado **marker** <br />
Este parametro nos permitira crear marcadores en los puntos que graficamos, por ejemplo: <br />
Si queremos colocar circulos en los puntos de interes podemos usar la funcion de esta forma <br />
**plt.plot([puntosX],[puntosY],marker = "o")** <br /n>

En general podemos definir algunos tipos de marcadores, mas informacion en el siguiente 
[LINK](https://www.w3schools.com/python/matplotlib_markers.asp)


```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16],marker = "*")    #Tambien podemos marcar los puntos con una estrella
```

Los markers especificados como String siguen el siguiente formato: **marker|line|color** <br />
Si deseamos omitir alguno solo no se coloca la especificacion
Ejm-> o-r (De esta forma decimos que el marcador sea un circulo de color rojo)<br />

La sintaxis para el tipo de linea es la siguiente
'-' 	Solid line	
':' 	Dotted line	
'--' 	Dashed line	
'-.' 	Dashed/dotted line

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16],"o-.r") 
```

```python
#Tambien podemos aumentar el tamanio de los marcadores con el parametro ms
plt.plot([1, 2, 3, 4], [1, 4, 9, 16],"o-.r",ms = 10) 
```
