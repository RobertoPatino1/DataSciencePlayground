```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Podemos colocar varios graficos al mismo tiempo en una ventana con la funcion subplot
#plt.subplots(1,2,1)
#El primer parametro representa las filas, el segundo las columnas y el tercero representa el 
#indice del plot. Por ejemplo
#Si queremos mostrar 2 graficos uno junto al otro:
plt.subplot(1,2,1) #Grafico 1
plt.plot([0, 1, 2, 3],[3, 8, 1, 10])

plt.subplot(1,2,2) #Grafico 2
plt.plot([0, 1, 2, 3],[10, 20, 30, 40])


plt.show()
```

```python
#Si quisieramos mostrar uno abajo del otro: 
plt.subplot(2,1,1) #Grafico 1
plt.plot([0, 1, 2, 3],[3, 8, 1, 10])

plt.subplot(2,1,2) #Grafico 2
plt.plot([0, 1, 2, 3],[10, 20, 30, 40])


plt.show()
```

```python
#En general podemos graficar cuantos plots deseemos y ponerle los titulos que les correspondan: 
plt.suptitle("MI NEGOCIO")
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 1)
plt.plot(x,y)
plt.title("Ventas")

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 2)
plt.plot(x,y)
plt.title("Ingresos")

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 3)
plt.plot(x,y)
plt.title("Egresos")

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 4)
plt.plot(x,y)
plt.title("Consumos internos")

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 5)
plt.plot(x,y)
plt.title("Consumos externos")

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 6)
plt.plot(x,y)
plt.title("Gastos")

plt.show()
```
