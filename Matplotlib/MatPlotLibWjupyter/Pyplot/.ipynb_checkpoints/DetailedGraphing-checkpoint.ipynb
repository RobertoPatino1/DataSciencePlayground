```python
import matplotlib.pyplot as plt
import numpy as np
#Podemos definir labels para los ejes X y Y de nuestras graficas junto con el titulo de la misma
```

```python
plt.title("Sports data",color = "red",size = 20)
plt.plot([80, 85, 90, 95, 100, 105, 110, 115, 120, 125],[240, 250, 260, 270, 280, 290, 300, 310, 320, 330],marker= "o",c = "red")
plt.xlabel("Pulso promedio",color = "blue")
plt.ylabel("Calorias quemadas",color = "blue")
plt.show()
```

```python
#Tambien podemos agregar un grid a nuestros graficos con la funcion grid
plt.grid()
plt.show()
```

```python
#O podemos hacer que las cuadriculas solo sean en un eje
plt.grid(axis = 'x')
plt.show()
plt.grid(axis = 'y')
plt.show()
```

```python
#Podemos especificar diferentes propiedades para las cuadriculas como son:
#Color, tipo de linea y grosor de linea
plt.grid(color = "green",linestyle = "--",linewidth = 0.5)
```
