```python
import matplotlib.pyplot as plt
import numpy as np
#matplotlib suele trabajar en conjunto con numpy
#El modulo para graficar es pyplot
```

```python
#PodemosCrear graficas con esta libreria!
#Dibujando una recta de un punto a otro
coordenadasX = np.array([0,6])
coordenadasY = np.array([0,250])
```

```python
#la funcion plot dibuja una linea de punto a punto
plt.plot(coordenadasX,coordenadasY)
plt.show()
```

```python
#Otro grafico
#El primer argumento pasado a plot corresponde a los puntos en x, el segundo a los puntos en y
plt.plot([1,2,3],[5,7])
plt.show()
```
